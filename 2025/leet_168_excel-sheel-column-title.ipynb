例如：

A -> 1
B -> 2
C -> 3
...
Z -> 26
AA -> 27
AB -> 28 
...
 

示例 1：

输入：columnNumber = 1
输出："A"
示例 2：

输入：columnNumber = 28
输出："AB"
示例 3：

输入：columnNumber = 701
输出："ZY"
示例 4：

输入：columnNumber = 2147483647
输出："FXSHRXW"
 

提示：

1 <= columnNumber <= 231 - 1

```python
class Solution:
    def convertToTitle(self, columnNumber: int) -> str:
        result = []
        while columnNumber > 0:
            columnNumber -= 1  # Decrement to make it 0-indexed
            current_char = chr(ord('A') + (columnNumber % 26))
            result.append(current_char)
            columnNumber //= 26
        return ''.join(reversed(result))
```

```python
def test(func):
    assert func(1) == "A"
    assert func(28) == "AB"
    assert func(701) == "ZY"
    assert func(52) == "AZ"
    assert func(703) == "AAA"
    assert func(2147483647) == "FXSHRXW"
    print("All tests passed.")
```

```python
s = Solution()
test(s.convertToTitle)
```

```python
def convertToNumer(str):
    res  = 0
    for c in str:
        res *= 26
        res += ord(c) - ord('A') + 1
    return res

def testConvertToNumer(func):
    assert func("A") == 1
    assert func("AB") == 28
    assert func("ZY") == 701
    assert func("AZ") == 52
    assert func("AAA") == 703
    assert func("FXSHRXW") == 2147483647
    print("All tests passed.")
testConvertToNumer(convertToNumer)
```

```python
def convertToTitle(columnNumber):
    digits = 6
    while columnNumber < 26 ** digits:
        digits -= 1
    result = ""
    while columnNumber > 0:
        c = columnNumber // (26**digits)
        result += chr(ord('A') + c - 1)
        columnNumber = columnNumber % (26**digits)
        digits -= 1
    return result

```

他不能从最大的开始。26**n 不仅是 A(N)*(26**n) 贡献的。 A(N-1)*(26**(N-1)) 也能贡献。 A(N-1)可以取到 26
所以只能从低位开始算。也是不停除以26，不过要对当前位置，A(N) == 26这个case 不停特判。




```python
def convertToTitle2(columnNumber):
    digits = 0
    result = ""
    while columnNumber > 0:
        c = columnNumber % 26
        columnNumber = columnNumber // 26
        if c == 0:
            c = 26
            columnNumber -= 1
        result += chr(ord("A") + c - 1)
    return "".join(reversed(result))
test(convertToTitle2)
```
